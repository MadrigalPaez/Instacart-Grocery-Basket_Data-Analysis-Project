## IF-STATEMENTS WITH USER-DEFINED FUNCTIONS TASK AND IF-STATEMENTS WITH THE LOC() FUNCTION

Creating a flag for pricing, with the following criteria:

- If the item’s price is lower than or equal to 5 -> “low-range product”
- If the item’s price is above 5 but lower than or equal to 15 -> “mid-range product”
- If the item’s price is above 15 -> “high-range product”

```python
#CREATING A SMALLER DATAFRAME

orders_and_products_combined_short = orders_and_products_combined[:1000000]
```

```python
#COMPARING DATASETS

orders_and_products_combined.shape
```

```python
orders_and_products_combined_short.shape
```

```python
#CREATING "price_label" FUNCTION

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
orders_and_products_combined_short['price_range'] = orders_and_products_combined_short.apply(price_label, axis=1)
```

```python
orders_and_products_combined_short.head()
```

```python
orders_and_products_combined_short['price_range'].value_counts(dropna = False)
```

```python
#APPLYING "price_label" FUNCTION W/ LOC

orders_and_products_combined_short.loc[orders_and_products_combined_short['prices'] > 15, 'price_range_loc'] = 'High-range product'
orders_and_products_combined_short.loc[(orders_and_products_combined_short['prices'] <= 15) & (orders_and_products_combined_short['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
orders_and_products_combined_short.loc[orders_and_products_combined_short['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

## Establish “Busiest day” instead of “Busiest days”. This label should correspond with the 2 busiest days of the week. Establish "Slowest day" with the 2 slowest days. Create a new column for this using a suitable method:

```python
#CHECKING FREQUENCY ON “order_day_of_week” COLUMN

orders_and_products_combined_short['order_day_of_week'].value_counts(dropna = False)
```

Busiest day = 0 and 1
Slowest day = 4 and 4

```python
#CREATING AN EMPTY LIST

result_busiest_and_slowest_day = []

for value in orders_and_products_combined_short["order_day_of_week"]:
    if value in [0, 1]:
        result_busiest_and_slowest_day.append("Busiest day")
    elif value in [3, 4]:
        result_busiest_and_slowest_day.append("Slowest day")
    else:
        result_busiest_and_slowest_day.append("Regularly busy")
```

```python
#COMBINING "result_busiest_and_slowest_day" WITH the "orders_and_products_combined_short" dataframe

orders_and_products_combined_short['busiest_day'] = result_busiest_and_slowest_day
```

## Identify the busiest hours of the day. Rather than by hour, periods of time labeled as: “Most orders”, “Average orders”, and “Fewest orders”. New column containing these labels called “busiest_period_of_day”:

```python
orders_and_products_combined_short.head()
```

```python
#CHECKING FREQUENCY ON “order_hour_of_day” COLUMN

orders_and_products_combined_short['order_hour_of_day'].value_counts(dropna = False)
```

```python
orders_and_products_combined_short['order_hour_of_day'].describe()
```

- 25% of the orders happen before 10
- 50% of the orders happen before 13
- 75% of the orders happen before 16

Division of periods of time:

- Most orders: >75000 orders -> From 9 to 16
- Average orders: Between 10000 and 80000 -> From 7 to 8, and from 17 to 23
- Fewest orders: <10000 orders -> From 0 to 6

```python
#CREATING AN EMPTY LIST

result_busiest_periods_of_day = []

for value in orders_and_products_combined_short["order_hour_of_day"]:
    if value in [9, 10, 11, 12, 13, 14, 15, 16]:
        result_busiest_periods_of_day.append("Most orders")
    elif value in [0, 1, 2, 3, 4, 5, 6]:
        result_busiest_periods_of_day.append("Average orders")
    else:
        result_busiest_periods_of_day.append("Fewest orders")
```

```python
result_busiest_periods_of_day
```

```python
#COMBINING "result_busiest_periods_of_day" WITH the "orders_and_products_combined_short" dataframe

orders_and_products_combined_short['busiest_period_of_day'] = result_busiest_periods_of_day
```
